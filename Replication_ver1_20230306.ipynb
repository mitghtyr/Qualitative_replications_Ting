## APM calculation and Hedging strategy

```python

class APM_Factor(object):
    def __init__(self,stock_name,index_name,onsite_date,past_date,lookback,df):
        self.n_stock = len(stock_name)-1
        self.stock_name = stock_name
        
        self.onsite_date = onsite_date
        self.past_date =past_date
        self.lookback = lookback
        self.df = df
        
    
        # past 20 days prices
        self.ind = (self.df['date']<self.onsite_date)&(self.df['date']>=self.past_date)
        self.price_seq = self.df.loc[self.ind,self.stock_name]
        self.time_seq = self.df.loc[self.ind,['time']]
    
    def AM_ret(self):
        # 过去20天上午收益
        open_price = self.price_seq.loc[self.time_seq['time']==931,:]
        open_price.reset_index(drop=True,inplace=True)
        
        close_price = self.price_seq.loc[self.time_seq['time']==1130,:]
        close_price.reset_index(drop=True,inplace=True)
        
        res = np.log(close_price/open_price)
        
        return res
    def PM_ret(self):
        # 过去20天下午收益
        open_price = self.price_seq.loc[self.time_seq['time']==1301,:]
        open_price.reset_index(drop=True,inplace=True)
        close_price = self.price_seq.loc[self.time_seq['time']==1500,:]
        close_price.reset_index(drop=True,inplace=True)
        res = np.log(close_price/open_price)
        
        return res
    
    
    def Ret20(self):
        open_ind = (self.df['date']==(self.past_date))
        open_price = self.df.loc[open_ind,self.stock_name[:(self.n_stock)]]
        open_price = open_price.iloc[-1]
        open_price.reset_index(drop=True,inplace=True)
        
        
        
        close_ind = (self.df['date']==(self.onsite_date))
        close_price = self.df.loc[close_ind,self.stock_name[:(self.n_stock)]]
        
        close_price = close_price.iloc[-1]
        close_price.reset_index(drop=True,inplace=True)
        
        
        res = np.log(close_price/open_price)
        return res
        
    
    def regress(self):
        
        am = self.AM_ret()
        pm = self.PM_ret()
        lm_df = pd.concat([am,pm])
        lm_df.reset_index(drop=True,inplace=True)
        lm_model = LinearRegression()
        lm_model.fit(lm_df.iloc[:,-1].values.reshape(-1,1),lm_df.iloc[:,0:self.n_stock].values.reshape(-1,self.n_stock))
        resid = lm_df.iloc[:,0:self.n_stock].values-\
            np.dot(np.ones([2*self.lookback,1]),lm_model.intercept_.reshape(1,-1))-\
            np.dot(lm_df.iloc[:,-1].values.reshape(-1,1),lm_model.coef_.reshape(1,-1))

        delta = (resid[:self.lookback]-resid[self.lookback:])
        delta = (np.mean(delta,axis=0) )/np.std(delta,axis=0)/np.sqrt(self.lookback)
        
        return delta
    
    def apm_factor(self):
        delta = self.regress()
        ret20 = self.Ret20()
        
        apm_model = LinearRegression( fit_intercept=False)
        apm_model.fit(ret20.values.reshape(-1,1),delta.reshape(-1,1))
        
        resid= delta.reshape(-1,1)-apm_model.coef_[0]*ret20.values.reshape(-1,1)
        return resid
    
    
```

```python
factor_seq_apm =pd.DataFrame(np.zeros([len(trade_period)-20,n_stocks]),columns=stock_name[:n_stocks])



lookback=20


for i in range((len(trade_period)-20)):
    trade_date = trade_period[i+lookback]
    trade_past = trade_period[i]

    try:
        return_test = APM_Factor(stock_name,index_name,trade_date,trade_past,lookback,df)
    
        apm_factor= return_test.apm_factor()
        factor_seq_apm.iloc[i,:] = apm_factor.reshape(1,-1)
    except:
        continue
    
    
    


```
